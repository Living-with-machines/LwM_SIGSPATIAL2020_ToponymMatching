```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
def standardize(foldername, output, dataset):
    df_csv = pd.read_csv(foldername + "/" + dataset, sep="\t", header=None, error_bad_lines=False)
    df_csv = df_csv.iloc[:,0:3]
    df = pd.DataFrame(df_csv.values, columns = ["Source", "Target", "Match"])
    df['Match'] = df['Match'].apply(lambda x: str(x).upper())
    df = df.sample(frac=1)
    train = df.reset_index(                  # need to keep the index as a column
        ).groupby('Match'                    # split by "group"
        ).apply(lambda x: x.sample(frac=0.9) # in each group, do the random split
        ).reset_index(drop=True              # index now is group id - reset it
        ).set_index('index')                 # reset the original index
    test = df.drop(train.index)              # now we can subtract it from the rest of data
    train = train.sample(frac=1)
    test = test.sample(frac=1)
    train.to_csv("../../datasets/intrinsic/" + output + "_trainval.txt", sep="\t", header=False, index=False)
    test.to_csv("../../datasets/intrinsic/" + output + "_test.txt", sep="\t", header=False, index=False)
```

```python
datasets = [("gb1900", "gb1900", "gb1900dataset.tsv"), 
            ("ocrdataset", "ocr", "ocr_posneg_whitespaces.tsv"), 
            ("santos", "santos", "dataset-string-similarity.txt"), 
            ("wikigaz", "wikigaz", "wikigaz_en_dataset.txt")]

for d in datasets:
    standardize(d[0], d[1], d[2])
```
