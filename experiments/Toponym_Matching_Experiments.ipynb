# Toponym Matching experiments

```python
from DeezyMatch import train as dm_train
from DeezyMatch import plot_log
from DeezyMatch import finetune as dm_finetune
from DeezyMatch import inference as dm_inference

from pathlib import Path
from shutil import copyfile
```

## Evaluation of Toponym Matching methods on different datasets (table 5)

### 1. Dataset: wikigaz_en

```python
if not Path('./models/wikigaz_en_l2norm_001').is_dir():
    # train a new model
    dm_train(input_file_path="./inputs/input_dfm.yaml",
             dataset_path="../datasets/toponym_matching_datasets/wikigaz_en_trainval.txt",
             model_name="wikigaz_en_l2norm_001")
```

```python
# plot log file
plot_log(path2log="./models/wikigaz_en_l2norm_001/log.txt", 
         output_name="wikigaz_en_l2norm_001")
```

**!!! WARNING:** copy best checkpoint model to final model

```python
best_checkpoint = "checkpoint00003"
```

```python
copyfile("./models/wikigaz_en_l2norm_001/" + best_checkpoint + ".model", "./models/wikigaz_en_l2norm_001/wikigaz_en_l2norm_001.model")
copyfile("./models/wikigaz_en_l2norm_001/" + best_checkpoint + ".model_state_dict", "./models/wikigaz_en_l2norm_001/wikigaz_en_l2norm_001.model_state_dict")
```

```python
# model inference using a model stored at pretrained_model_path and pretrained_vocab_path 
dm_inference(input_file_path="./models/wikigaz_en_l2norm_001/input_dfm.yaml",
             dataset_path="../datasets/toponym_matching_datasets/wikigaz_en_test.txt",
             pretrained_model_path="./models/wikigaz_en_l2norm_001/wikigaz_en_l2norm_001.model",
             pretrained_vocab_path="./models/wikigaz_en_l2norm_001/wikigaz_en_l2norm_001.vocab")
```
